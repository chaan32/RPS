```python
from ultralytics import YOLO

model = YOLO("model/yolo/yolo11s-pose.pt") 
results = model.track(source=0, show=True, conf=0.25,persist=True, tracker="custom_botsort.yaml")
# source=1 : 웹캠으로 실시간 추적
# show=True : 추적 결과를 화면에 표시
# conf=0.25 : 신뢰도 임계값 설정 (0.25 이상인 경우에만 추적 결과를 표시)
# persist=True : 추적 결과를 지속적으로 유지 (웹캠이 켜져 있는 동안 추적 결과가 계속 표시됨)
```
