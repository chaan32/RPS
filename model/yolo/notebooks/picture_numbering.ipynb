```python
from pathlib import Path
from PIL import Image
from pillow_heif import register_heif_opener
register_heif_opener()

# 원본 사진들이 들어있는 폴더
input_dir = Path(r"C:\Users\niced\Downloads\forklift_image_night\before")

# 결과 저장 폴더
output_dir = Path(r"C:\Users\niced\Downloads\forklift_image_night\after")
output_dir.mkdir(exist_ok=True)

# 처리할 확장자들
extensions = {".png", ".jpg", ".jpeg", ".bmp", ".webp", ".heic"}

# 파일 모으기
image_files = [f for f in input_dir.iterdir() if f.suffix.lower() in extensions]
image_files.sort()  # 이름순 정렬

failed = []
success_count = 0

for img_path in image_files:
    try:
        img = Image.open(img_path).convert("RGB")  # JPG로 저장하려면 RGB로 변환
        success_count += 1
        save_path = output_dir / f"{success_count}.jpg"
        img.save(save_path, "JPEG", quality=95)
    except Exception as e:
        print(f"실패: {img_path.name} / {e}")
        failed.append(img_path.name)
        continue

print("완료")
print(f"성공 파일 수: {success_count}")
print(f"실패 파일 수: {len(failed)}")
print("실패 파일 목록:", failed)
```
